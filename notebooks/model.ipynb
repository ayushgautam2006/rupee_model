```python
bankwise = bankwise.drop(columns=['AED_INR', 'IDR_INR'])

exchange_rate = exchange_rate.ffill()

forex_reserves = forex_reserves.ffill()

foreign_trade = foreign_trade.drop(columns=[
    'Imports_INR_Cr', 'Imports_USD_Mn',
    'Trade_Balance_INR_Cr', 'Trade_Balance_USD_Mn',
    'NonOil_Balance_INR_Cr', 'NonOil_Balance_USD_Mn'
])

major_price_indices = major_price_indices.drop(columns=['CPI_IW_2016'])
major_price_indices = major_price_indices.ffill()

# Verify
for name, df in [("Crude Oil", crude_oil), ("BankWise", bankwise), 
                 ("Exchange Rate", exchange_rate), ("Forex Reserves", forex_reserves),
                 ("Foreign Trade", foreign_trade), ("Major Price Indices", major_price_indices)]:
    print(f"{name}: {df.isnull().sum().sum()} empty cells")
```

```python
# Filter data from 2014-2024
crude_oil = crude_oil[(crude_oil['Date'].dt.year >= 2014) & (crude_oil['Date'].dt.year <= 2024)]
bankwise = bankwise[(bankwise['Date'].dt.year >= 2014) & (bankwise['Date'].dt.year <= 2024)]
exchange_rate = exchange_rate[(exchange_rate['Date'].dt.year >= 2014) & (exchange_rate['Date'].dt.year <= 2024)]
forex_reserves = forex_reserves[(forex_reserves['Date'].dt.year >= 2014) & (forex_reserves['Date'].dt.year <= 2024)]
foreign_trade = foreign_trade[(foreign_trade['Month'].dt.year >= 2014) & (foreign_trade['Month'].dt.year <= 2024)]
major_price_indices = major_price_indices[(major_price_indices['Date'].dt.year >= 2014) & (major_price_indices['Date'].dt.year <= 2024)]

print("Filtered data (2014-2024):")
for name, df in [("Crude Oil", crude_oil), ("BankWise", bankwise), 
                 ("Exchange Rate", exchange_rate), ("Forex Reserves", forex_reserves),
                 ("Foreign Trade", foreign_trade), ("Major Price Indices", major_price_indices)]:
    print(f"  {name}: {df.shape[0]} rows")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax1 = plt.subplots(figsize=(14, 5))
ax1.plot(exchange_rate['Date'], exchange_rate['USD_INR_RBI'], color='royalblue', label='USD/INR')
ax1.set_ylabel('INR per USD', color='royalblue')
ax2 = ax1.twinx()
ax2.plot(crude_oil['Date'], crude_oil['Crude_Price'], color='orange', alpha=0.6, label='Crude Oil')
ax2.set_ylabel('Crude Oil Price (USD)', color='orange')
plt.title('USD/INR vs Crude Oil Price')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.tight_layout()
plt.show()
```

```python
usd_monthly = exchange_rate.set_index('Date')['USD_INR_RBI'].resample('MS').mean()
crude_monthly = crude_oil.set_index('Date')['Crude_Price'].resample('MS').mean()
forex_monthly = forex_reserves.set_index('Date')['Total_Reserves_USD_Mn'].resample('MS').mean()
trade_monthly = foreign_trade.set_index('Month')['Exports_USD_Mn'].resample('MS').mean()
cpi_monthly = major_price_indices.set_index('Date')['CPI24_Combined'].resample('MS').mean()
wpi_monthly = major_price_indices.set_index('Date')['WPI_All'].resample('MS').mean()

# Merge all into one dataframe
merged = pd.DataFrame({
    'USD_INR': usd_monthly,
    'Crude_Price': crude_monthly,
    'Forex_Reserves_USD_Mn': forex_monthly,
    'Exports_USD_Mn': trade_monthly,
    'CPI': cpi_monthly,
    'WPI': wpi_monthly
}).dropna()
print("Merged dataset shape:", merged.shape)
print(f"Date range: {merged.index.min().date()} → {merged.index.max().date()}")
print(merged.head())
```

```python
corr = merged.corr()['USD_INR'].drop('USD_INR').sort_values()

plt.figure(figsize=(8, 5))
colors = ['red' if x < 0 else 'green' for x in corr]
corr.plot(kind='barh', color=colors)
plt.title('Correlation of Each Feature with USD/INR', fontsize=13)
plt.xlabel('Correlation Coefficient')
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 7))
sns.heatmap(merged.corr(), annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, linewidths=0.5, square=True)
plt.title('Correlation Heatmap — All Features', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
print("\n════════════════════════════════════")
print("         KEY INSIGHTS               ")
print("════════════════════════════════════")
corr_all = merged.corr()['USD_INR'].drop('USD_INR').sort_values()
for feature, val in corr_all.items():
    direction = "↑ INR weakens" if val > 0 else "↓ INR strengthens"
    strength = "Strong" if abs(val) > 0.6 else "Moderate" if abs(val) > 0.3 else "Weak"
    print(f"{feature:<25} r={val:+.2f}  [{strength}] — {direction}")
print("\nNote: r closer to +1 or -1 = stronger relationship with USD/INR")
```
