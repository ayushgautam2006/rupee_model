```python
for name, df in [("Crude Oil", crude_oil), ("BankWise", bankwise), 
                 ("Exchange Rate", exchange_rate), ("Forex Reserves", forex_reserves),
                 ("Foreign Trade", foreign_trade), ("Major Price Indices", major_price_indices)]:
    print(f"\n── {name} ──")
    total_rows = len(df)
    for col in df.columns:
        null_count = df[col].isnull().sum()
        if null_count > 0:
            pct = (null_count / total_rows) * 100
            print(f"  {col}: {null_count} missing ({pct:.1f}%)")
```

```python
bankwise = bankwise.drop(columns=['AED_INR', 'IDR_INR'])

exchange_rate = exchange_rate.ffill()

forex_reserves = forex_reserves.ffill()

foreign_trade = foreign_trade.drop(columns=[
    'Imports_INR_Cr', 'Imports_USD_Mn',
    'Trade_Balance_INR_Cr', 'Trade_Balance_USD_Mn',
    'NonOil_Balance_INR_Cr', 'NonOil_Balance_USD_Mn'
])

major_price_indices = major_price_indices.drop(columns=['CPI_IW_2016'])
major_price_indices = major_price_indices.ffill()

# Verify
for name, df in [("Crude Oil", crude_oil), ("BankWise", bankwise), 
                 ("Exchange Rate", exchange_rate), ("Forex Reserves", forex_reserves),
                 ("Foreign Trade", foreign_trade), ("Major Price Indices", major_price_indices)]:
    print(f"{name}: {df.isnull().sum().sum()} empty cells")
```
